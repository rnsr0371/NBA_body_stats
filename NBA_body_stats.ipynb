```python
#必要なライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression
```

```python
#データのインポート
Players=pd.read_csv("NBA_data.csv")
```

```python
#身長をフィートからセンチに変換
feet=Players["Height"].str.split("-")
metre=[]

for i in range(len(feet)):
    feet[i][0]=int(feet[i][0])
    feet[i][1]=int(feet[i][1])
    
    metre.append((feet[i][0]*0.3048+feet[i][1]*0.3048/10)*100)
    
Players["Height(cm)"]=pd.Series(metre)
```

```python
#体重をオンスからKgに変換
kg=[]
for i in Players["Weight"]:
    i=int(i)
    kg.append(i*0.4535)
Players["Weight(kg)"]=pd.Series(kg)
```

```python
#BMIを計算
bmi=[]
for h,w in zip(Players["Height(cm)"],Players["Weight(kg)"]):
    bmi.append(w/((h/100)**2))
    
Players["BMI"]=pd.Series(bmi)
```

```python
#要約統計量の確認
Players.loc[:,["Age","Height(cm)","Weight(kg)","BMI","PTS","REB","AST"]].describe()
```

```python
#BMIとAgeの相関を見る
plt.plot(Players["BMI"],Players["Age"],"o")
plt.title("Correlation between BMI and Age")
plt.xlabel("BMI")
plt.ylabel("Age")
plt.show()
print("相関係数は"+str(round(Players["BMI"].corr(Players["Age"]),3)))
```

```python
#BMIとPTSの相関を見る
plt.plot(Players["BMI"],Players["PTS"],"o")
plt.title("Correlation between BMI and PTS")
plt.xlabel("BMI")
plt.ylabel("PTS")
plt.show()
print("相関係数は"+str(round(Players["BMI"].corr(Players["PTS"]),3)))
```

```python
#BMIとASTの相関を見る
plt.plot(Players["BMI"],Players["AST"],"o")
plt.title("Correlation between BMI and AST")
plt.xlabel("BMI")
plt.ylabel("AST")
plt.show()
print("相関係数は"+str(round(Players["BMI"].corr(Players["AST"]),3)))
```

```python
#BMIとREBの相関を見る
plt.plot(Players["BMI"],Players["REB"],"o")
plt.title("Correlation between BMI and REB")
plt.xlabel("BMI")
plt.ylabel("REB")
plt.show()
print("相関係数は"+str(round(Players["BMI"].corr(Players["REB"]),3)))
```

```python
#BMIとPTSの相関を見る
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
map1=ax1.scatter(Players["BMI"],Players["PTS"],c=Players["Height(cm)"],s=15)
fig.colorbar(map1,ax=ax1)
plt.title("Correlation between BMI and PTS")
plt.xlabel("BMI")
plt.ylabel("PTS")

#BMIとREBの相関
ax2=fig.add_subplot(1,3,2)
map2=ax2.scatter(Players["BMI"],Players["REB"],c=Players["Height(cm)"],s=15)
fig.colorbar(map2,ax=ax2)
plt.title("Correlation between BMI and REB")
plt.xlabel("BMI")
plt.ylabel("REB")

#BMIとASTの相関
ax3=fig.add_subplot(1,3,3)
map3=ax3.scatter(Players["BMI"],Players["AST"],c=Players["Height(cm)"],s=15)
fig.colorbar(map3,ax=ax3)
plt.title("Correlation between BMI and AST")
plt.xlabel("BMI")
plt.ylabel("AST")

plt.show()
```
